```python
import os

from langchain.agents import Tool
from langchain.agents import load_tools

from crewai import Agent, Task, Process, Crew
from langchain.llms import Ollama
ollama_llm = Ollama(model="openhermes")
```

```python


# Enter your research topic 
research_topic= " Phd on Large Language Model and Human computer interaction"

from langchain.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()
"""
1. Create agent who works for you as a researcher. Researcher will access information availble over internet to about the topic.
2. Create one more agent who writes about the the content provided by the researcher. 
3. Crete one more agent as review who will review the content and improvise it . Define agents for resardefine agents that are going to research latest  tools and write a blog about it 

"""
job_seeker = Agent(
    role="Phd Student",
    goal="Find the list of latest Phd topics on which include large langauge model and human computer interaction internet.",
    backstory="""You are an student who is looking for Phd topic in the field of large langauge model and human computer interaction.""",
    verbose=True,
    allow_delegation=False,
    tools=[search_tool],
    llm=ollama_llm
)

writer = Agent(
    role="Critics",
    goal="To criticize the topic why the provided each topic is not suitable or feasibility for Phd and what are the disadvantages and future advantages. Also, in case of topic is too valuable should also state that.",
    backstory="""You are professor in university and suggest phd topics to students""",
    verbose=True,
    allow_delegation=True,
    llm=ollama_llm
)
reviewer = Agent(
    role="Writer",
    goal="Review the content and summarize the each idea ",
    backstory="""You are expert reviewer with 10 years of exprience in reviewing phd topics. You are also a scientist in Large langauge model and human machine interface.""",
    verbose=True,
    allow_delegation=True,
    llm=ollama_llm
)

task_report = Task(
    description="""Conduct a thorough examination of the latest phd topics in large language model, large action model, and Human computer interaction. 
    Identify the key trends, breakthrough technologies, and potential industry impacts.
    Your final product should be a list of top 10 phd topics. 
    """,
    agent= job_seeker,
)

task_blog = Task(
    description="""Identify significant disadvantage and advantages of the provided phd topics, in case of future scope, feasibilty.""",
    agent=writer,
)

task_critique = Task(
    description="""Summarize the result of provided information.
    """,
    agent= reviewer,
)

# instantiate crew of agents
crew = Crew(
    agents=[job_seeker, writer, reviewer],
    tasks=[task_report, task_blog, task_critique],
    verbose=2,
    process=Process.sequential,  # Sequential process will have tasks executed one after the other and the outcome of the previous one is passed as extra content into this next.
)

# Get your crew to work!
result = crew.kickoff()

print("######################")
print(result)
```
