```python
import os
from crewai import Agent, Task, Crew, Process

from langchain.llms import Ollama
ollama_llm = Ollama(model="mistral",base_url="http://10.36.88.133:11434")

from langchain.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()
import fitz
```

```python

def read_pdf(file_path):
    try:
        # Open the PDF file
        with fitz.open(file_path) as pdf_document:
            text = ""
            # Iterate through each page
            for page_number in range(pdf_document.page_count):
                # Get the page
                page = pdf_document[page_number]
                # Extract text from the page
                text += page.get_text()

            return text
    except Exception as e:
        print(f"Error reading PDF: {e}")
        return None
    
def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except Exception as e:
        print(f"Error reading the text file: {e}")
        return None
    
def write_to_notepad(content, file_path):
    try:
        # Try to open the file in write mode
        with open(file_path, 'w') as file:
            # Write the content to the file
            file.write(content)
        print(f"Content successfully written to {file_path}")
    except FileNotFoundError:
        # If the file is not found, create it and write the content
        with open(file_path, 'w') as file:
            file.write(content)
        print(f"File {file_path} not found. Created and content written.")
    except Exception as e:
        # Handle other exceptions
        print(f"An error occurred: {e}")
```

# Search Job and create interview questions

```python
backstory="""You have worked as automation and manual tester in robot framework at a leading tech company.
  Your expertise lies in designing test automation framework.
  you also have experience in project release management."""
city='stuttgart'
```

```python

# Define your agents with roles and goals
job_seeker = Agent(
  role='Job seeker',
  goal='Finding job in germany',
  backstory= backstory,
  verbose=True,
  allow_delegation=False,
  tools=[search_tool],
  llm=ollama_llm
)
writer = Agent(
  role='Interviewer',
  goal='Generate interview questions as per the job',
  backstory="""You are a renowned interviewer of automation teseting known for
  hiring candidates.""",
  verbose=True,
  allow_delegation=True,
  llm=ollama_llm
)

# Create tasks for your agents
task1 = Task(
  description=f"""Find the top 10 jobs advertisement in automation testing in robot framework in {city} city. 
  Collect the name of the company, URL of the job, and job requirements.""",
  agent=job_seeker
)

task2 = Task(
  description="""Using the information provided, develop 10 interview questions for each job advertisement""",
  agent=writer
)

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[job_seeker, writer],
  tasks=[task1, task2],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)

# Get your crew to work!
result = crew.kickoff()

print("######################")
print(result)
```

```python


# Define your agents with roles and goals
writer = Agent(
  role='Writer',
  goal="""Based on the information of job seeking candidate and company requirement provided to, write a convincing cover letter for the candidate. So the person can get the job.""",
  backstory=""" You are a writer and human resource manager, who helps in writing conver letter for the candidate""",
  verbose=False,
  allow_delegation=False,
  tools=[search_tool],
  llm=ollama_llm
)
reviewer = Agent(
  role='Reviewer',
  goal='Review the cover letter provided to you, check for spelling mistakes and also simplify the language in the cover letter.',
  backstory="""You are expert in reviewing the candidate cover letter and improving them.""",
  verbose=False,
  allow_delegation=True,
  llm=ollama_llm
)

# Create tasks for your agents
task1 = Task(
  description=f"""First understand  the candidate profile {condidate_profile} and company requirements {company_requirements}. Then, You have to write a cover letter 
  in simple, interesting, and convicing way. Where cover letter should showcase candidate skills matching with the company requirements. 
  Write it in simple english language with minimal jargons. 
  """,
  agent=writer
)

task2 = Task(
  description=f"""Review the provided cover letter, and improvise the cover letter if required. 
  The cover letter should showcase candidate skills are matching with company requriements i.e {company_requirements}. Look for spelling mistakes, jargons, and formatting.
  Improvise the given cover letter and provide a new cover letter.
  """,
  agent=reviewer
)

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[writer, reviewer],
  tasks=[task1, task2],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)

# Get your crew to work!
cover_letter = crew.kickoff()
print(cover_letter)

write_to_notepad(cover_letter, "result/cover_letter.txt")
```
